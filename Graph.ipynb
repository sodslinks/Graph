````python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

sns.set()
````

````python
df = sns.load_dataset("planets")
````

````python
sns.set(style = "darkgrid")
````

# seabornでの散布図
```relplot```
    relationship plot

`x`
    x軸の変量

`y`
    y軸の変量


`data`
    データフレーム

`hue`
    マーカーの色分け

`style`
    マーカーの形状、変数(列名)で指定

`size`
    マーカーのサイズ。変数(列名)で指定


`legend`
    
 
`kind`
    scatter、lineのいずれか

`height`
    グラフの高さ。インチ

`aspect`
    グラフの幅。高さに対する比率


````python
# seabornでの散布図
sns.relplot(x = "distance", y = "mass", hue = "year",  size = "mass" ,data = df)
````

## カテゴリごとにプロット

`col`
    カテゴリのカラムを指定。カテゴリごとに横に並べる
    
`row`
    カテゴリのカラムを指定。カテゴリごとに縦に並べる

````python
sns.relplot("distance", y = "mass", col = "number", data = df)
````

````python
sns.relplot(x = "distance", y = "mass", col = "year", row = "method", data = df)
````

````python
df1 = sns.load_dataset("fmri")
df1
````

## `relplot`のライン

````python
sns.relplot(x = "timepoint", y = "signal", hue = "subject",
            col = "region", row = "event", kind = "line", data = df1)
````
